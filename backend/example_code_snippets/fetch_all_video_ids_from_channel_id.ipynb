# Fetches ALL (public) video ids (and some other random details like publish time) given channel id

```python
from dotenv import load_dotenv
import os

SMALL_CHANNEL_ID = 'UC8VRaOMul3RCkc6-3qL54UQ'
ANOTHERROOF_ID = "UCHEnZhUKjZSLYs3jJ0raKZA" # This is a larger YouTube channel (only 45 videos at the time lol)

load_dotenv()
api_key = os.getenv("YOUTUBE_API_KEY")
```

```python
from googleapiclient.discovery import build
import json

videos_data = []

with build('youtube', 'v3', developerKey=api_key) as youtube:
    next_page_token = None

    while True:
        request = youtube.search().list(
            part="id,snippet",
            channelId=ANOTHERROOF_ID,
            maxResults=50,
            pageToken=next_page_token,
            type="video"
        )
        
        response = request.execute()
        videos_data.extend(response['items'])

        if not next_page_token:
            break


print('Number of videos: ', len(videos_data))
print(json.dumps(videos_data, indent=4))
```
